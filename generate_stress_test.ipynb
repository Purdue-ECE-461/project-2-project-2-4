```python
import logging
import threading
import time
import concurrent.futures
import requests
```

```python
response_times_ms_list = []
result_list = []
def thread_function(name):
    # logging.info("Thread %s: starting", name)
    result = requests.get("https://ece-461-project-2-team-4.uc.r.appspot.com/package/package6")    # time.sleep(10)
    response_times_ms_list.append(result.elapsed.total_seconds() *1000)
    # logging.info("Thread %s: finishing", name)

if __name__ == "__main__":
    format = "%(asctime)s.%(msecs)03d: %(message)s"
    logging.basicConfig(format=format, level=logging.INFO,
                        datefmt="%H:%M:%S")

    with concurrent.futures.ThreadPoolExecutor(max_workers=10000) as executor:
        executor.map(thread_function, range(10000))

    print(len(response_times_ms_list))
    
```
